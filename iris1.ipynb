```python
# Import the necessary libraries
import pandas as pd
import seaborn as sns

# Load the Iris dataset using seaborn
data = sns.load_dataset('iris')

# Display the first few rows of the dataset to ensure it has been loaded correctly
data.head()
```

```python
# Input features: selecting the petal length and petal width columns
x = data.iloc[:, [2, 3]].values

# Output labels: selecting the species column
y = data.iloc[:, 4].values
```

```python
# Import the train_test_split function from sklearn
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets, with 25% of the data used for testing
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
# Import the StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler and fit it to the training data
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)

# Transform the test data using the same scaler
xtest = sc_x.transform(xtest)

# Print the first 10 rows of the scaled training data
print(xtrain[0:10, :])
```

```python
# Import the Gaussian Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB

# Create an instance of the GaussianNB classifier and fit it to the training data
classifier = GaussianNB()
classifier.fit(xtrain, ytrain)
```

```python
# Predict the labels for the test data
y_pred = classifier.predict(xtest)
```

```python
# Import the confusion_matrix function from sklearn
from sklearn.metrics import confusion_matrix

# Compute and print the confusion matrix to evaluate the model's performance
cm = confusion_matrix(ytest, y_pred)
print("Confusion Matrix : \n", cm)
```

```python
# Import the accuracy_score function from sklearn
from sklearn.metrics import accuracy_score

# Compute and print the accuracy of the model
print("Accuracy : ", accuracy_score(ytest, y_pred))
```
